```python
import requests
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
# Load latest UCC database from Zenodo
zenodo_url = requests.get('https://doi.org/10.5281/zenodo.8250523', allow_redirects=True).url
df_ucc = pd.read_csv(zenodo_url + "/files/UCC_cat.csv", dtype={"Blue_str": str})
```
